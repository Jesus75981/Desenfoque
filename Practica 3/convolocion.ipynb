```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def aplicar_desenfoque_selectivo(img, centro, ancho, alto):
    img_copia = np.copy(img)
    val_x, val_y = centro
    mitad_ancho = ancho // 2
    mitad_alto = alto // 2

    mascara = np.zeros_like(img, dtype=np.uint8)
    cv2.ellipse(mascara, (val_x, val_y), (mitad_ancho, mitad_alto), 0, 0, 360, (255, 255, 255), -1)
    
    
    mascara_invertida = cv2.bitwise_not(mascara)

    img_desenfocada = cv2.GaussianBlur(img, (51, 51), 0)
    
    img_copia[mascara_invertida != 0] = img_desenfocada[mascara_invertida != 0]
    
    return img_copia

imagen = cv2.imread('JudeBellingham.jpeg')
imagen_rgb = cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB)


imagen_desenfocada = aplicar_desenfoque_selectivo(imagen_rgb, (150, 100), 100, 50)


plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(imagen_rgb)
plt.title('Imagen Original')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(imagen_desenfocada)
plt.title('Imagen con Desenfoque Selectivo')
plt.axis('off')

plt.show()

```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

original = cv2.imread("JudeBellingham.jpeg")
original_rgb = cv2.cvtColor(original, cv2.COLOR_BGR2RGB)

gris = cv2.cvtColor(original, cv2.COLOR_BGR2GRAY)
gauss = cv2.GaussianBlur(gris, (5, 5), 0)
canny = cv2.Canny(gauss, 50, 150)

contornos, _ = cv2.findContours(canny.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

mask = np.zeros_like(gris)
cv2.drawContours(mask, contornos, -1, (255), thickness=cv2.FILLED)

persona_enfocada = cv2.bitwise_and(original, original, mask=mask)

fondo_desenfocado = cv2.GaussianBlur(original, (15, 15), 0)
fondo_desenfocado = cv2.bitwise_and(fondo_desenfocado, fondo_desenfocado, mask=~mask)

imagen_final = cv2.add(persona_enfocada, fondo_desenfocado)

plt.figure(figsize=(12, 8))
plt.imshow(cv2.cvtColor(imagen_final, cv2.COLOR_BGR2RGB))
plt.title('Persona Enfocada con Fondo Desenfocado')
plt.axis('off')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('Motoveloz.jpeg')

kernels = {
    'Identidad': np.array([[0, 0, 0], [0, 1, 0], [0, 0, 0]]),
    'Suavizado': np.ones((3, 3), np.float32) / 9,
    'Realce de bordes': np.array([[0, -1, 0], [-1, 5, -1], [0, -1, 0]]),
    'Deteccin de bordes': np.array([[-1, -1, -1], [-1, 8, -1], [-1, -1, -1]]),
    'Enfoque': np.array([[0, -1, 0], [-1, 5, -1], [0, -1, 0]])
}

plt.figure(figsize=(15, 10))
columns = 3
rows = len(kernels) // columns + 1
for i, (name, kernel) in enumerate(kernels.items(), 1):
    plt.subplot(rows, columns, i)
    img_convolved = cv2.filter2D(img, -1, kernel)
    plt.imshow(cv2.cvtColor(img_convolved, cv2.COLOR_BGR2RGB))
    plt.title(name)
    plt.axis('off')

plt.subplot(rows, columns, len(kernels) + 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Original')
plt.axis('off')

plt.tight_layout()
plt.show()
```
